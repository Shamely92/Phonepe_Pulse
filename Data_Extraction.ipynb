```python
import streamlit as st
from PIL import Image
import os
import json
from streamlit_option_menu import option_menu
import pandas as pd
import sqlite3
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
import warnings
import mysql.connector as sql

```

```python
path="C:/Users/THEBEST/Desktop/Jupyter workbook/pulse/data/aggregated/transaction/country/india/state/"
agg_trans_list=os.listdir(path)
agg_trans_list

```

```python
#Dataframe of aggregated Transactions

#This is to extract the data's to create a dataframe
column1 = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],
            'Transaction_amount': []}
for i in agg_trans_list:
    p_i = path + i + "/"
    agg_year_list = os.listdir(p_i)
    
    for j in agg_year_list:
        p_j = p_i + j + "/"
        agg_file_list = os.listdir(p_j)
        
        for k in agg_file_list:
            p_k = p_j + k
            data = open(p_k, 'r')
            A = json.load(data)
            
            for z in A['data']['transactionData']:
                name = z['name']
                count = z['paymentInstruments'][0]['count']
                amount = z['paymentInstruments'][0]['amount']
                column1['Transaction_type'].append(name)
                column1['Transaction_count'].append(count)
                column1['Transaction_amount'].append(amount)
                column1['State'].append(i)
                column1['Year'].append(j)
                column1['Quarter'].append(int(k.strip('.json')))
                
df_agg_trans = pd.DataFrame(column1)
```

```python
#Dataframe of aggregated user

path2 = "C:/Users/THEBEST/Desktop/Jupyter workbook/pulse/data/aggregated/user/country/india/state/"
agg_user_list = os.listdir(path2)

columns2 = {'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Count': [],
            'Percentage': []}
for i in agg_user_list:
    p_i = path2 + i + "/"
    agg_year_list = os.listdir(p_i)
    
    for j in agg_year_list:
        p_j = p_i + j + "/"
        agg_file_list = os.listdir(p_j)

        for k in agg_file_list:
            p_k = p_j + k
            data = open(p_k, 'r')
            B = json.load(data)
            try:
                for z in B["data"]["usersByDevice"]:
                    brand_name = z["brand"]
                    counts = z["count"]
                    percents = z["percentage"]
                    columns2["Brands"].append(brand_name)
                    columns2["Count"].append(counts)
                    columns2["Percentage"].append(percents)
                    columns2["State"].append(i)
                    columns2["Year"].append(j)
                    columns2["Quarter"].append(int(k.strip('.json')))
            except:
                pass
df_agg_user = pd.DataFrame(columns2)
```

```python
#Dataframe of map transactions
path3 = "C:/Users/THEBEST/Desktop/Jupyter workbook/pulse/data/map/transaction/hover/country/india/state/"

map_trans_list = os.listdir(path3)

columns3 = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [],
            'Amount': []}

for i in map_trans_list:
    p_i = path3 + i + "/"
    map_year_list = os.listdir(p_i)
    
    for j in map_year_list:
        p_j = p_i + j + "/"
        map_file_list = os.listdir(p_j)
        
        for k in map_file_list:
            p_k = p_j + k
            data = open(p_k, 'r')
            C = json.load(data)
            
            for z in C["data"]["hoverDataList"]:
                district = z["name"]
                count = z["metric"][0]["count"]
                amount = z["metric"][0]["amount"]
                columns3["District"].append(district)
                columns3["Count"].append(count)
                columns3["Amount"].append(amount)
                columns3['State'].append(i)
                columns3['Year'].append(j)
                columns3['Quarter'].append(int(k.strip('.json')))
                
df_map_trans = pd.DataFrame(columns3)
```

```python
#Dataframe of map user
path4 = "C:/Users/THEBEST/Desktop/Jupyter workbook/pulse/data/map/user/hover/country/india/state/"

map_user_list = os.listdir(path4)

columns4 = {"State": [], "Year": [], "Quarter": [], "District": [], "RegisteredUser": [], "AppOpens": []}

for i in map_user_list:
    p_i = path4 + i + "/"
    map_year_list = os.listdir(p_i)
    
    for j in map_year_list:
        p_j = p_i + j + "/"
        map_file_list = os.listdir(p_j)
        
        for k in map_file_list:
            p_k = p_j + k
            data = open(p_k, 'r')
            D = json.load(data)
            
            for z in D["data"]["hoverData"].items():
                district = z[0]
                registereduser = z[1]["registeredUsers"]
                appOpens = z[1]['appOpens']
                columns4["District"].append(district)
                columns4["RegisteredUser"].append(registereduser)
                columns4["AppOpens"].append(appOpens)
                columns4['State'].append(i)
                columns4['Year'].append(j)
                columns4['Quarter'].append(int(k.strip('.json')))
                
df_map_user = pd.DataFrame(columns4)
```

```python
#Dataframe of top transactions
path5 = "C:/Users/THEBEST/Desktop/Jupyter workbook/pulse/data/top/transaction/country/india/state/"

top_trans_list = os.listdir(path5)
columns5 = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Transaction_count': [], 'Transaction_amount': []}

for i in top_trans_list:
    p_i = path5 + i + "/"
    top_year_list = os.listdir(p_i)
    
    for j in top_year_list:
        p_j = p_i + j + "/"
        top_file_list = os.listdir(p_j)
        
        for k in top_file_list:
            p_k = p_j + k
            data = open(p_k, 'r')
            E = json.load(data)
            
            for z in E['data']['pincodes']:
                name = z['entityName']
                count = z['metric']['count']
                amount = z['metric']['amount']
                columns5['Pincode'].append(name)
                columns5['Transaction_count'].append(count)
                columns5['Transaction_amount'].append(amount)
                columns5['State'].append(i)
                columns5['Year'].append(j)
                columns5['Quarter'].append(int(k.strip('.json')))
df_top_trans = pd.DataFrame(columns5)
```
